```python
# Importar librer√≠as necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
# Cargar los datos generados desde el archivo CSV
df = pd.read_csv('data.csv')
```

```python

# Crear gr√°ficos de Gasto vs CTR y Gasto vs CPA
plt.figure(figsize=(12, 5))

# Gasto vs CTR
plt.subplot(1, 3, 1)
sns.scatterplot(x='Gasto (‚Ç¨)', y='CTR', hue='Plataforma', data=df, alpha=0.7)
plt.title('Gasto (‚Ç¨) vs CTR')

# Gasto vs CPA
plt.subplot(1, 3, 2)
sns.scatterplot(x='Gasto (‚Ç¨)', y='CPA', hue='Plataforma', data=df, alpha=0.7)
plt.title('Gasto (‚Ç¨) vs CPA')

# Gasto vs CR
plt.subplot(1, 3, 3)
sns.scatterplot(x='Gasto (‚Ç¨)', y='CR', hue='Plataforma', data=df, alpha=0.7)
plt.title('Gasto (‚Ç¨) vs CR')

plt.tight_layout()
plt.show()
```

```python
# Visualizar la distribuci√≥n del gasto
plt.figure(figsize=(6, 4))
sns.histplot(df['Gasto (‚Ç¨)'], bins=30, kde=True, color='blue')
plt.title('Distribuci√≥n del Gasto (‚Ç¨)')
plt.xlabel('Gasto (‚Ç¨)')
plt.ylabel('Frecuencia')
plt.tight_layout()
plt.show()
```

```python
# Crear gr√°ficos de dispersi√≥n (scatter plots) para ver las relaciones entre variables clave
plt.figure(figsize=(12, 10))

# Gasto vs Impresiones
plt.subplot(2, 2, 1)
sns.scatterplot(x='Gasto (‚Ç¨)', y='Impresiones', hue='Plataforma', data=df, alpha=0.7)
plt.title('Gasto (‚Ç¨) vs Impresiones')

# Gasto vs Conversiones
plt.subplot(2, 2, 2)
sns.scatterplot(x='Gasto (‚Ç¨)', y='Conversiones', hue='Plataforma', data=df, alpha=0.7)
plt.title('Gasto (‚Ç¨) vs Conversiones')

# Impresiones vs Conversiones
plt.subplot(2, 2, 3)
sns.scatterplot(x='Impresiones', y='Conversiones', hue='Plataforma', data=df, alpha=0.7)
plt.title('Impresiones vs Conversiones')

plt.tight_layout()
plt.show()

# Comparaci√≥n entre plataformas con boxplots
plt.figure(figsize=(14, 10))

# Boxplot de Gasto por plataforma
plt.subplot(2, 2, 1)
sns.boxplot(x='Plataforma', y='Gasto (‚Ç¨)', data=df)
plt.title('Distribuci√≥n del Gasto (‚Ç¨) por Plataforma')

# Boxplot de Impresiones por plataforma
plt.subplot(2, 2, 2)
sns.boxplot(x='Plataforma', y='Impresiones', data=df)
plt.title('Distribuci√≥n de Impresiones por Plataforma')

# Boxplot de Clics por plataforma
plt.subplot(2, 2, 3)
sns.boxplot(x='Plataforma', y='Clics', data=df)
plt.title('Distribuci√≥n de Clics por Plataforma')

# Boxplot de Conversiones por plataforma
plt.subplot(2, 2, 4)
sns.boxplot(x='Plataforma', y='Conversiones', data=df)
plt.title('Distribuci√≥n de Conversiones por Plataforma')

plt.tight_layout()
plt.show()
```
